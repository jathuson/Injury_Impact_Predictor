```python
from nba_api.stats.endpoints import LeagueDashLineups
import pandas

lineups = LeagueDashLineups(season='2022-23', season_type_all_star='Regular Season', per_mode_detailed='Totals', group_quantity=5)

database = lineups.lineups.get_data_frame()

database.describe()
```

```python
database.to_csv('lineups_2022_2023.csv',index = False)
```

```python
columns_list = database.columns.tolist()

print(columns_list)
```

```python
database.loc[:,'GROUP_SET':'PLUS_MINUS'].head()
```

```python
team_abbr = "MIL"  # Team abbreviation to filter by
df_team = database[(database["TEAM_ABBREVIATION"] == team_abbr) & (database["MIN"]> 100)].sort_values(by=["PLUS_MINUS"], ascending=False)   # Filter the DataFrame by team abbreviation
df_team.loc[:,'GROUP_SET':'PLUS_MINUS'].head()
```
